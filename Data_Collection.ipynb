### DATASET 1 - Game releases for 2023

```python
url = "https://en.wikipedia.org/wiki/2023_in_video_games"
page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')

tables = soup.find_all('table')

print(len(tables))
```
